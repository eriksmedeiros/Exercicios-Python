**Testando combina√ß√µes**

Seth tinha uma forma peculiar de guardar suas senhas de n√∫meros.

Para cada senha, ele cria um conjunto de n√∫meros e armazena uma soma.

Para recuperar a senha, basta identificar qual par de n√∫meros gera a soma guardada.

Depois, √© s√≥ ordenar o par.

Escreva um c√≥digo para ajud√°-lo, considerando que Seth vai te informar a soma e em seguida 10 n√∫meros.

```python
def valordasenha(soma_valor, n):
    pares = []

    for i in range(len(n)):
        for x in range(i + 1, len(n)):
            if n[i] + n[x] == soma_valor:
                par = (n[i], n[x])
                pares.append(par)

    if len(pares) == 0:
        return None

    pares.sort()
    senha = f"{pares[0][0]}{pares[0][1]}"
    return(senha)

soma_valor = int(input("Insira a soma: "))
n = []
for i in range(10):
    numero = int(input(f"Digite o {i+1}¬∫ n√∫mero: "))
    n.append(numero)

senha = valordasenha(soma_valor, n)
2
print("Sua Senha √©:", senha)
```

**Peda√ßo de senhas**

Zac tem um m√©todo diferente para gerar suas senhas.

Em vez de escolher a senha, ele escolhe uma frase que seja f√°cil de lembrar.

Dentro dessa frase, ele identifica qual o maior trecho que n√£o cont√©m letras repetidas, e usa esse trecho como senha.

Escreva um c√≥digo para ajudar Zac a gerar uma senha, considerando que ele vai te informar uma frase.



```python
frase = input("digite a frase: ").replace(" ","")
teste = ""
maior = ""

for x in frase:
    if x not in teste:
        teste += x
    else:
        if len(teste) > len(maior):
            maior = teste
        teste = x
if len(teste) > len(maior):
    maior = teste


print(maior)
```
